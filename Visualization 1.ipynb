```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.plot(area,price)
plt.show()
```

```python
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(area,price)
plt.xlim((200,2000))
plt.ylim((5,50))
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(area,price)
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

## Multiple Plots

```python
x = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()#to plot two graphs
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(labels=["sqr","log"])
plt.show()# to give names
```

```python
plt.figure()
plt.plot(x,y1,color="black")
plt.plot(x,y2,color="green")
plt.legend(labels=["sqr","log"])
plt.show()#to change colour
```

## Histograms

```python
df = pd.read_csv("cars.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure()
plt.hist(df["width"])
plt.title("Width Distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.hist(df["width"],bins=5)# to change bin counts
plt.title("Width Distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
```

```python
mean = df["width"].mean()
median = df["width"].median()
plt.figure()
plt.hist(df["width"])
plt.axvline(mean,color="yellow")
plt.axvline(median,color="red")
plt.legend(labels=["mean","median"])
plt.title("Width and Frequency Distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
```
