## Visualization 2 (Seaborn)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cars.csv")
```

```python
df.head()
```

```python
x = np.arange(20)
y = np.sqrt(x)
```

```python
sns.lineplot(x,y)
plt.show()
```

```python
sns.distplot(df["width"])
plt.show()
```

```python
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
sns.countplot(data=df, y="make")
plt.show()
```

```python
sns.countplot(data=df, x="body-style",hue="fuel-type")
plt.show()
```

```python
sns.countplot(data=df, x="drive-wheels",hue="fuel-type")
plt.show()
```

## Heatmap

```python
cor = df.corr()
```

```python
cor
```

```python
sns.heatmap(cor)
plt.show()
```

```python
sns.heatmap(cor, annot=True)
plt.show()
```

```python
sns.heatmap(cor, annot=True, cmap="viridis")
plt.show()
```

### Scattered plot

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x="width", y="price",hue="body-style")
plt.show()
```

## Sub Plots

```python

fig,axes = plt.subplots(2,2,figsize=(8,6))

axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)

plt.show()
```

```python
fig, axes =plt.subplots(2,2,figsize=(15,15))
sns.boxplot(data=df, x="price", y="body-style", ax=axes[0,0])
sns.heatmap(cor, annot=True, cmap="viridis",ax=axes[0,1])
sns.scatterplot(data=df, x="width", y="price",hue="body-style",ax=axes[1,0])
sns.countplot(data=df, x="body-style",hue="fuel-type",ax=axes[1,1])

plt.show()
```
